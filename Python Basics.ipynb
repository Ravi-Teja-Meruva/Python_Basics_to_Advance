### Scope of Variables: Local and Gloabl Variables

##### Let's define a global variable, "name" outside the function. We will return its value using a function "my_function" and see that we would be able to access its value using that function also.

```python
name = 'variable outside function'

def my_function():
    return name
```

```python
print(my_function())
print(name)
```

##### So, we were able to access the value of a global variable from the inside the function. Let's see what will happen if we try to change the value of global variable inside the function

```python
name = "variable outside function"

def my_function():
    name = "variable inside function"
    return name
```

```python
print(my_function())
print(name)
```

##### Let's see how can we update the value of a global variable inside any user_defined function.

```python
name = 'variable outside the function'

def my_function():
    global name
    name = 'variable inside the function'
    return name
```

```python
print(my_function())
print(name)
```

### Recursion

```python
def count_hand_shakes(total_persons_in_room):
    
    count = 0
    
    for i in range(1, total_persons_in_room):
        for j in range(i+1,total_persons_in_room+1):
            print(f'Handshake of {i} and {j}')
            count = count +1
            
    print(f'Total handshakes {count}')
```

```python
count_hand_shakes(total_persons_in_room = 5)
```

```python
def count_hand_shakes_recursion(total_persons_in_room):
    
    if total_persons_in_room == 2:
        return 1 
    else:
        return count_hand_shakes_recursion(total_persons_in_room - 1) + (total_persons_in_room-1)
```

```python
count_hand_shakes_recursion(5)
```

#### Example 1 : Creating Class and Object in Python

```python
class Parrot:
    # attributes
    species = "bird"
    # instance of attributes
    def __init__(self,name,age):
        self.name = name
        self.age = age
        
# creating an instance of the class
blu = Parrot("Blu",10)
woo = Parrot("Woo",15)
```

```python
# access the instance attributes
print(f"{blu.name} is {blu.age} years old")
print(f"{woo.name} is {woo.age} years old")
```

### Methods

Methods are functions defined inside the body of a class. They are used to define the behaviors of an object.

#### Example 2 : Creating Methods in Python

```python
class Parrot:
    
    def __init__(self,name,age):
        self.name = name
        self.age = age
    # instance methods
    def sing(self,song):
        return f'{self.name} sings {song} song'
    
    def dance(self):
        return f'{self.name} is now dancing'
```

```python
# instantiate the object
blu = Parrot("Blu",10)

# call our instance methods
print(blu.sing("'Butabomma'"))
print(blu.dance())
```

## Inheritance

It is a way of creating a new class by using details of an existing class without modifying it. The newly  formed class is a derived class(or child class). Similarly, the existing class is a base class(or parent class).

#### Example 3 : Use of Inheritance in Python

```python
# parent class

class Bird:
    
    def __init__(self):
        print("Bird is ready")
    def whoisThis(self):
        print("Bird")
    def swim(self):
        print("Swim faster")
        
        
# child class

class Penguin(Bird):
    
    def __init__(self):
        super().__init__()
        print("Penguin is ready")
    def whoisThis(self):
        print("Penguin")
    def run(self):
        print("Run Faster")
```

```python
peggy = Penguin()
peggy.whoisThis()
peggy.swim()
```

## Polymorphism

It is an ability to use a common interface for multiple forms. Suppose, we need to color a shape, there are multiple shape options(rectangle,square,circle etc.) However we could use the same method to color any shape. This concept is called Polymorphism.

#### Example 5 : Using Polymorphism in Python 

```python
class Parrot:
    def fly(self):
        print("Parrot can fly")
    def swim(self):
        print("Parrot can't swim")
        
class Penguin:
    def fly(self):
        print("Penguin can't fly")
    def swim(self):
        print("Penguin can swim")
# common interface

def flying_test(bird):
    bird.fly()
    
#Instantiate Objects

blu = Parrot()
peggy = Penguin()

#passing the object
flying_test(blu)
flying_test(peggy)
```

### Multiple Inheritance

```python
# when method is overridden in both classes 

class Class1:
    def mi(self):
        print("In class 1")
class Class2(Class1):
    def mi(self):
        print("In Class 2")

class Class3(Class1):
    def mi(self):
        print("In class 3")
class Class4(Class2,Class3):
    pass
obj = Class4()
obj.mi()
        
```

```python
# when method is overriden only  in one of the classes 

class Class1:
    def mi(self):
        print("In class 1")
class Class2(Class1):
    pass

class Class3(Class1):
    def mi(self):
        print("In class 3")
class Class4(Class2,Class3):
    pass
obj = Class4()
obj.mi()
```

```python
# when we try to call the method m for class1
# class2, class3 from the method m of class4
class Class1:
    def mi(self):
        print("In class 1")
class Class2(Class1):
    def mi(self):
        print("In Class 2")

class Class3(Class1):
    def mi(self):
        print("In class 3")
class Class4(Class2,Class3):
    def mi(self):
        print("Im in class 4")
        Class2.mi(self)
        Class3.mi(self)
        Class1.mi(self)
obj = Class4()
obj.mi()
```

```python
class Car:
    def __init__(self,make,model):
        self.make = make
        self.model = model
    def __repr__(self):
        return f'<Car {self.make} {self.model}>'
```

```python
class garage:
    def __init__(self):
        self.cars= []
    def __len__(self):
        return len(self.cars)
    def add_car(self,car):
        if not isinstance(car,Car):
            raise TypeError(f'Tried to add a {car.__class__.__name__} to garage, but you can add only Car Objects')
        self.cars.append(car)
```

```python
ford = garage()
ford.add_car('Ecosport')
```

```python
ford = garage()
Fiesta = Car('Ford','Fiesta')
ford.add_car(Fiesta)
print(len(ford.cars))
```
